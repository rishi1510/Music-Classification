```python
from scipy.io import wavfile

def trim_wav( originalWavPath, newWavPath , start, end ):
    '''
    :param originalWavPath: the path to the source wav file
    :param newWavPath: output wav file * can be same path as original
    :param start: time in seconds
    :param end: time in seconds
    :return:
    '''
    sampleRate, waveData = wavfile.read( originalWavPath )
    startSample = int( start * sampleRate )
    endSample = int( end * sampleRate )
    wavfile.write( newWavPath, sampleRate, waveData[startSample:endSample])
 
 
wp = r"output.wav"
trim_wav(wp, wp.replace(".wav", "_trimmed.wav"), 55,85)
```

```python
import librosa
import librosa.display

M = librosa.feature.melspectrogram(y, sr, 
                                           fmax = sr/2, # Maximum frequency to be used on the on the MEL scale
                                           n_fft=2048, 
                                           hop_length=512, 
                                           n_mels = 96, # Set as per the Google Large-scale audio CNN paper
                                           power = 2) # Power = 2 refers to squared amplitude
```

```python
import pylab
from matplotlib import cm

log_power = librosa.power_to_db(M, ref=np.max)# Covert to dB (log) scale

# Plotting the spectrogram
pylab.figure(figsize=(5,5))
pylab.axis('off') 
pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge

librosa.display.specshow(log_power, cmap=cm.jet)
pylab.savefig('1.jpg', bbox_inches=None, pad_inches=0)
plt.show()
```
